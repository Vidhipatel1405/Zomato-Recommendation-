```python
import seaborn as sns
```

```python
cuisine_count=df['cuisines'].value_counts().head(10)
```

```python
cuisine_count.plot(kind='bar',title='top 10 cuisines')
```

```python
sns.histplot(df['rate'],bins=20,kde=True)
plt.title('Distribution of Ratings')
plt.show()
```

```python
df['approx_cost(for two people)'] = df['approx_cost(for two people)'].clip(upper=df['approx_cost(for two people)'].quantile(0.95))
sns.boxplot(df['approx_cost(for two people)'])
plt.title("distribution for approx_cost")
plt.show()
```

```python
from sklearn.preprocessing import MinMaxScaler
```

```python
scaler=MinMaxScaler()
df['normalized_rate'] = scaler.fit_transform(df[['rate']])
```

```python
#Correlation Analysis
corr=df[['rate','votes','approx_cost(for two people)']].corr()
sns.heatmap(corr,annot=True,cmap='coolwarm')
plt.title('Feature Correlation')
plt.show()
```

```python
#convert values into numeric
df['rate']=df['rate'].replace('/5','').astype(float)
```

```python
#df['approx_cost(for two people)'] = df['approx_cost(for two people)'].str.replace(',', '').astype(float)
```

```python
#location based analysis
location_counts=df['location'].value_counts().head(10)
location_counts.plot(kind='bar', title='top locations by restaurant count')
plt.show()
```

```python
avg_ratings=df.groupby('location')['rate'].mean().sort_values(ascending=False).head(10)
```

```python
avg_ratings.plot(kind='bar',title='Top locations by average ratings')
```

```python
#top rated cuisines
cuisine_ratings = df.groupby('cuisines')['rate'].mean().sort_values(ascending=False).head(10)
cuisine_ratings.plot(kind='bar', title='Top rated cuisines')
plt.show()
```

```python
df['listed_in(type)'].value_counts().plot(kind='pie', autopct='%1.1f%%', title='Restaurant Types')
plt.show()
```

```python
df.groupby('listed_in(city)')['rate'].mean().sort_values(ascending=False).head(10).plot(kind='bar', title='Top Cities by Average Ratings')
plt.show()
```
